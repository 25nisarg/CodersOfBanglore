Collect raw instagram data of all openai followers

Answer these questions:

```python
with open("initialdata.txt", encoding="utf-8") as f:
    data=f.read()
```

```python
data
```

```python
chunks=data.split("\n\n")
```

```python
chunks[0]
```

```python
def parse_chunk(chunk):
    chunk=chunk.strip()
    sep_chunk= chunk.split("\n")
    username=sep_chunk[0]
    no_of_post=sep_chunk[1]
    no_of_followers=sep_chunk[2]
    no_of_following=sep_chunk[3]
    name=sep_chunk[4]
    type=sep_chunk[5]
    bio="\n".join(sep_chunk[6:])
    print(username, no_of_post,no_of_followers, no_of_following,name,type,bio,sep="\n")
    return{"username":username,"no_of_post": no_of_post,"no_of_followers": no_of_followers,"no_of_following": no_of_following,"name":name,"type":type,"bio":bio}

parse_chunk(chunks[0])
```
